## histplot()

```python
# tips["total_bill"] : refers to a column like df["total_bill"]
# bins = number of bars in the graph
# count = 0 to 10 range me approximately 8 bill generated
# kernal density estimate = generate Smooth curve 

sns.histplot(tips["total_bill"], kde = True, bins = 10)
```

## boxplot()

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)                  # day wise
```

```python
sns.boxplot(x = 'sex', y = 'total_bill', data = tips)                 # gender wise
```

```python
# hue : separate data per day and create compareson 
sns.boxplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', hue = 'smoker', data = tips)
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
```

## scatterplot()

```python
# Add Style, hue, size

sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, style='time', hue='day', size='size')
```

## lineplot()

```python
# give range of payment 
sns.lineplot(x = 'size', y = 'total_bill', data = tips)
```

## countplot()

```python
# count number of total bills per day (means count rows per day)
sns.countplot(y = 'day', data = tips)
```

```python
sns.countplot(y = 'day', hue = 'smoker', data = tips)
```

## barplot()

```python
# barplot() generate average of that day
# black line shows uncertanity and error (every time when we run this code the main bar vary in that range)
sns.barplot(x='day', y='total_bill', data=tips)
```

## pairplot()

```python
# create all combination of all graphs (numaric data)
sns.pairplot(hue = 'sex', data=tips)
```
