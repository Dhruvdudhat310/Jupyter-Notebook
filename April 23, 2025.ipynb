## histogram()

```python
import matplotlib.pyplot as plt
```

```python
# 1. What is the distribution of the maths score among the students?

plt.hist(df['math score'], bins=10)
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
# 2. How do the average maths score differ between male and female students?

plt.bar(df['gender'], df['math score'], data=df, color = ['Red','Green'])
```
