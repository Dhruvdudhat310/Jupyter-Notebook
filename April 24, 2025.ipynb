## Matplotlib and Seaborn Questions

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("students_performance.csv")
```

```python
# 1. What is the distribution of the maths score among the students?

plt.hist(df['math score'], bins=10)
plt.title('Distribution of maths score among the students')
plt.xlabel('Maths Score')
plt.ylabel('Number of Students')
plt.grid(axis='y')
plt.show()
```

```python
# 2. How do the average maths score differ between male and female students?

plt.bar(df['gender'], df['math score'], color=['red','green'])
```

```python
# 3. Is there a relation between completing the test preparation course and maths course?
# (We can use bar or box plot)

plt.bar(df['test preparation'], df['math score'], color=['Yellow',"Red"])
plt.xlabel('Test Preparation')
plt.ylabel('Math Score')
```

```python
# 4. How are the reading and writing score related to each other?

sns.scatterplot(x='reading score', y='writing score', data=df )
```

```python
# 5. Dose parantal level of education seems to influence the maths score?
# (we can use Bar or boxplot )

plt.bar(df['parental level of education'], df['math score'], data = df)
```

```python
# 6. What is the distribution of reading score for each race/ethnicity column?

sns.scatterplot(x='race/ethnicity' ,y='reading score', data = df)
```

## NEW DATASET QUESTIONS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('employee_performance_dataset.csv')
```

```python
df
```

```python
# 1. Plot a chart of a employee age vs experience years.

plt.scatter(x='Age', y='Experience_Years', data=df, color='blue')
plt.title('Employee Age VS Experience Years')
plt.xlabel('Age')
plt.ylabel('Experience Years')
plt.grid(True)
plt.show()
```

```python
# 2. Show the percentage age of employee in each department.  

df['Department'].value_counts().mul(100).plot(kind='bar', color='red')
plt.xlabel('Departments')
plt.ylabel('Percentages')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# 3. Create a chart of performance rating for each employee.

sns.barplot(x='Name', y='Performance_Rating', data=df)
plt.title('Performance Rating for Each Employee')
plt.xlabel('Names')
plt.ylabel('Performance Ratings')
plt.xticks(rotation = 45)
plt.ylim(0,5)
plt.grid(axis='y')
plt.show()
```

```python
# 4. Weekly Working hours for all employee.

sns.barplot(x='Name', y='Weekly_Hours', data=df)
plt.title('Weekly Working Hours for All Employees')
plt.xlabel('Names')
plt.ylabel('Weekly Hours')
plt.xticks(rotation = 45)
plt.grid(axis = 'y')
plt.show()
```

```python
# 5. Subplot: (A) Weekly hours by Department (B) Performance rating by gender

fig, axes = plt.subplots(1, 2, figsize=(14,6))

sns.boxplot(x='Department', y='Weekly_Hours', hue='Department', data=df, ax=axes[0], palette='Set3')
axes[0].set_title('Weekly Hours by Department')

sns.boxplot(x='Gender', y='Performance_Rating', hue='Gender', data=df, ax=axes[1], palette='pastel')
axes[1].set_title('Performance Rating by Gender')

plt.tight_layout()
plt.show()
```

```python
# 6. Plot the average performance rating for each department.

df.groupby('Department')['Performance_Rating'].mean().plot(kind='bar', color='coral')
plt.title('Average Performance Rating by Department')
plt.xlabel('Departments')
plt.xticks(rotation = 45)
plt.ylabel('Average Rating')
plt.grid(axis='y')
plt.show()
```

```python
# 7. Visulize the distribution of weekly working hours across different genders.

sns.boxplot(x='Gender', y='Weekly_Hours', data=df)
plt.title('Distribution of weekly working Hours by Gender')
plt.grid(True)
plt.show()
```

```python
# 8. Compare the performance rating distribution by department.

sns.violinplot(x='Department', y='Performance_Rating', data=df)
plt.title('Performance Rating Distribution by Department')
plt.xlabel('Departments')
plt.ylabel('Performance_Rating')
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```

```python
# 9. Analyse the relationship between experience and performance rating.

sns.regplot(x='Experience_Years', y='Performance_Rating', data=df, color='orange')
plt.title('Experience VS Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Experience Years')
plt.grid(True)
plt.show()
```

```python
# 10. Compare weekly hours vs performance split by department.

sns.scatterplot(x='Weekly_Hours', y='Performance_Rating', hue='Department', data=df)
plt.title('Weekly Hours VS Performance by Department')
plt.xlabel('Weekly hours')
plt.ylabel('Performance Rating')
plt.grid(True)
```

```python
# 11. show pair-wise ralationship between age, weekly hours, experience year, performance.

sns.pairplot(df[['Age', 'Weekly_Hours', 'Experience_Years', 'Performance_Rating']])
plt.show()
```

```python
# 12. create correlation heatmap for numeric fields.

sns.heatmap(df[['Age', 'Weekly_Hours','Experience_Years','Performance_Rating']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
